```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
wine_rating_df = pd.read_csv(
    Path("winequality_red.csv")
)

wine_rating_df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2, random_state=1)

model
```

```python
model.fit(wine_rating_df)
```

```python
wine_ratings = model.predict(wine_rating_df)

print(wine_ratings)
```

```python
wine_rating_predictions_df = wine_rating_df.copy()

wine_rating_predictions_df['quality'] = wine_ratings

wine_rating_predictions_df.head()
```

```python
wine_rating_predictions_df.hvplot.scatter(
    x="free sulfur dioxide",
    y="total sulfur dioxide",
    by="quality"
)
```
